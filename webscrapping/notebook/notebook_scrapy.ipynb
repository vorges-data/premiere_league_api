# Imports

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

# Scrapping

```python
url_base = "https://www.transfermarkt.com.br/premier-league/marktwerte/wettbewerb/GB1/pos//detailpos/0/altersklasse/alle/land_id/0/plus/1"


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
response = requests.get(url_base, headers=headers)
soup = BeautifulSoup(response.text , 'html.parser')
```

```python
tabela = soup.find_all('table')[1]
```

### Classe HTML Odd

```python
rows = tabela.find_all('tr', attrs={'class':'odd'})

data = []

for row in rows: 
    data_rows = row.findAll('td')

    # Obter atributos específicos dentro das tags img
    img_data_src = row.find('img', {'class': 'bilderrahmen-fixed'}).get('data-src')
    img_src_flag = row.find('img', {'class': 'flaggenrahmen'}).get('src')
    img_src_club = row.find('img', {'class': ''}).get('src')
    club_name = row.find_all('a', {'title': True})[1].get('title')
    
    data_rows = [ele.text.strip() for ele in data_rows] # trazendo os textos dentro das tags
    data_row = [ele for ele in data_rows if ele]

    # Adicionar os atributos específicos à lista de dados
    data_row.extend([img_data_src, img_src_flag, img_src_club, club_name])
    
    data.append(data_row)

# Criar DataFrame
df = pd.DataFrame(data)
df
```

# Transformação

### Concatenando os Dfs da classe Even e Odd

```python
df_final = pd.concat([df_odd, df_even], ignore_index=True)
```

```python
df_final.head()
```

### Renomear as colunas

```python
df_final.columns = ['ranking', 'name_and_position', 'player_name', 'position', 'age', 'best_value_career', 'last_update', 'market_value', 'player_image', 'nation_image', 'club_image', 'club_name']
```

```python
df_final.head(2)
```

### Remover coluna name_and_position

```python
df_final = df_final.drop("name_and_position", axis=1, errors="ignore")
```

```python
df_final.head(1)
```

### Separar unidade nas colunas: best_value_career e market_value

```python
# Regex
def extrair_dados(valor):
    match = re.match(r'(\d+,\d+) mi. €', valor)
    if match:
        return match.group(1)
    else:
        return None


# Aplicar a função para extrair dados em ambas as colunas
df_final['best_value_career'] = df_final['best_value_career'].apply(extrair_dados)
df_final['market_value'] = df_final['market_value'].apply(extrair_dados)

# Criar as colunas moeda
df_final['currency_best_value_career'] = 'mi. €'
df_final['currency_market_value'] = 'mi. €'
```

```python
df_final.head(1)
```

### Reordenar as colunas

```python
# Lista com a ordem desejada das colunas
new_columns = ['ranking', 'player_name', 'position', 'age', 'club_name', 'best_value_career', 'currency_best_value_career', 'market_value', 'currency_market_value', 'player_image', 'nation_image', 'club_image', 'last_update']

# Reordenar as colunas
df_final = df_final[new_columns]
```

```python
df_final.head(1)
```

### Tipo de Dados

```python
def transformar_colunas(df):

    # Antes da conversão alterar a vírgula por ponto nos valores
    df['best_value_career'] = df['best_value_career'].str.replace(',','.')
    df['market_value'] = df['market_value'].str.replace(',','.')

    
    df['ranking'] = pd.to_numeric(df['ranking'])
    df['age'] = pd.to_numeric(df['age'])
    df['best_value_career'] = pd.to_numeric(df['best_value_career'])
    df['market_value'] = pd.to_numeric(df['market_value'])
    df['last_update'] = pd.to_datetime(df['last_update'], format='%d/%m/%Y')

    return df


df_final = transformar_colunas(df_final)
```

### Sort Dataframe pelo ranking

```python
df_final = df_final.sort_values('ranking', ascending=True).reset_index(drop=True)
```
